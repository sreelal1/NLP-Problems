# Stemming

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
stemmer = PorterStemmer()
```

```python
#to see all the stop words in english
stopwords.words('english')
```

```python
#here is the code to remove the stop words, then will do the stemming process

for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words)
```

# Bag of words

```python
#cleaning the text
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
ps = PorterStemmer()
wordnet = WordNetLemmatizer()
sentences = nltk.sent_tokenize(paragraph)
```

```python
#doing stemming
corpus = []
for i in range(len(sentences)):
    #in below text replacing all symbols, numbers etc with a space using re
    review = re.sub('[^a-zA-Z]',' ', sentences[i])
    #then doing stemming after preprocessing
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
    
```

```python
corpus
```

```python
#lets do lemmetization in above steps to understand the diff
sentences = nltk.sent_tokenize(paragraph)
corpus = []
for i in range(len(sentences)):
    #in below text replacing all symbols, numbers etc with a space using re
    review = re.sub('[^a-zA-Z]',' ', sentences[i])
    #then doing stemming after preprocessing
    review = review.lower()
    review = review.split()
    review = [wordnet.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
#Now creating BOW

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
X = cv.fit_transform(corpus).toarray() #converting to array to see it properly
```

# TF IDF

```python
# the initial steps are same.
# we are using lemmatization instead of stemming here bacause we need xact word


sentences = nltk.sent_tokenize(paragraph)
corpus = []

for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]',' ', sentences[i])
    review = review.lower()
    review = review.split()
    review = [wordnet.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# creating TF IDF model using TF IDF vectorizer

from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer()
Y = cv.fit_transform(corpus).toarray()
```
